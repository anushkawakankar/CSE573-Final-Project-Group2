```python
import pandas as pd
import numpy as np
import os
```

```python
dataset_path = "../DATA/ml-100k"
```

Load movie data

```python
movies_metadata = {"path":os.path.join(dataset_path, "u.item"), "header":["movie id","movie title","release date","video release date","IMDb URL","unknown","Action","Adventure","Animation","Children's","Comedy","Crime","Documentary","Drama","Fantasy","Film-Noir","Horror","Musical","Mystery","Romance","Sci-Fi","Thriller","War","Western"]}
movies=pd.read_table(movies_metadata["path"], encoding='latin-1', delimiter="|", header=None, names=movies_metadata["header"])
```

Load genre data


```python
genre_metadata = {"path":os.path.join(dataset_path, "u.genre"), "header":["genre","id"]}
genre=pd.read_table(genre_metadata["path"], encoding='latin-1', delimiter="|", header=None, names=genre_metadata["header"])
```

Load user data

```python
user_metadata = {"path":os.path.join(dataset_path, "u.user"), "header":["user id","age","gender","occupation","zip code"]}
users=pd.read_table(user_metadata["path"], encoding='latin-1', delimiter="|", header=None, names=user_metadata["header"])
```

Function to load rating data

```python
all_data_name = "u.data"

def load_ratings(file_name):
  metadata = {"path":os.path.join(dataset_path, file_name), "header":["user id","item id","rating","timestamp"]}
  return pd.read_table(metadata["path"], encoding='latin-1', delimiter="\t", header=None, names=metadata["header"])
```

```python
all_data = load_ratings(all_data_name)
all_data["item id"].value_counts()
```
